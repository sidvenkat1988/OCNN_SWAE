```
def rescale_data(file):
  data = np.load(file, allow_pickle=True)
  data = data.astype('float32')/255.0
  data = data.reshape((data.shape[0], data.shape[1], data.shape[2], 1))
  return data
```

```
positive_data = rescale_data('positive_images_28.npy')
```

```
positive_data.shape
```

```
plt.imshow(positive_data[0].reshape(28,28), cmap = 'gray')
```

```
negative_data = rescale_data('negative_images_28.npy')
```

```
negative_data.shape
```

```
index = np.random.permutation(len(negative_data))
```

```
negative_data = negative_data[index]
positive_data = positive_data[index]
```

```
X_train = negative_data[:16000]
X_test_negative = negative_data[16000:]
X_test_positive_balanced = positive_data[:4000]
X_test_positive_imbalanced = positive_data[4000:4500]
```

```
X_test_balanced = np.concatenate([X_test_negative, X_test_positive_balanced])
X_test_imbalanced = np.concatenate([X_test_negative, X_test_positive_imbalanced])
```
