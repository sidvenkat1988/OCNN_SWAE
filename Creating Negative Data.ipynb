```
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
len(os.listdir(os.getcwd()))
```

```
for i in os.listdir(os.getcwd()):
  if '.jpg' in i:
    pass
  else:
    print(i)
```

```
negative_images = [i for i in os.listdir(os.getcwd()) if '.jpg' in i]
```

```
negative_images_28 = []
for img in negative_images:
  image = cv2.imread(img)
  image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  image = cv2.resize(image, (28,28))
  negative_images_28.append(image)
  if len(negative_images_28)%100 == 0:
    print(len(negative_images_28))
```
