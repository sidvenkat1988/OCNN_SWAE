```
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
positive_images = [i for i in os.listdir(os.getcwd()) if '.jpg' in i]
```

```
len(positive_images)
```

```
original_positive_images = []
for img in positive_images:
  original_positive_images.append(cv2.imread(img))
  if len(original_positive_images)%500 == 0:
    print(len(original_positive_images))
```

```
positive_images_28 = []
for img in original_positive_images:
  img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  img = cv2.resize(img, (28, 28))
  positive_images_28.append(img)
  if len(positive_images_28)%500 == 0:
    print(len(positive_images_28))
np.save('positive_images_28', positive_images_28, allow_pickle=True)
```
