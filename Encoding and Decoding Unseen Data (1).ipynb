```
def encode_decode(data, trained_model_1, trained_model_2):
  model_1 = load_model(trained_model_1)
  encoded_data = model_1.predict(data)
  model_2 = load_model(trained_model_2)
  decoded_data = model_2.predict(encoded_data)
  return encoded_data, decoded_data 
```
